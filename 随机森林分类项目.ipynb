```python
import pandas as pd
import numpy as np
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn import model_selection, metrics
from sklearn.model_selection import train_test_split
```

```python
data_new=pd.read_csv(r'train_data.csv')
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_new.loc[:,'Existing_EMI':'Source_2'], 
                                                    data_new['Disbursed'], test_size=0.2, random_state=422)
```

```python
RFC=RandomForestClassifier(n_estimators=100).fit(X_train,y_train)
print(RFC.get_params)
print(RFC.score(X_train,y_train))
RFC.score(X_test,y_test)
```

```python
para_dict={'n_estimators':[50],'max_depth':[24],'criterion':['entropy','gini'],'class_weight':['balanced','balanced_subsample'],
           'min_samples_split':[4],'max_features':[24],'max_leaf_nodes':[178]}
GCV=GridSearchCV(estimator = RandomForestClassifier(oob_score=True,random_state=10,min_samples_leaf=1 ),param_grid=para_dict,cv=2)
```

```python
GCV.fit(X_train,y_train)
```

```python
GCV.best_params_, GCV.best_score_
```

```python
OptimalRFC=RandomForestClassifier(class_weight='balanced_subsample',criterion='entropy',max_depth=24,max_features=24,max_leaf_nodes=900,
min_samples_split=4,n_estimators=50,oob_score=False,random_state=19070,min_samples_leaf=1, min_impurity_decrease=0.000000).fit(X_train,y_train)
print(OptimalRFC.get_params)
print(OptimalRFC.score(X_train,y_train))
RFC.score(X_test,y_test)
```

```python
data1=pd.read_csv('bank-full.csv',delimiter=';')
```

```python
from sklearn import preprocessing
for feature in ['job','marital','education','default','housing','loan','contact','month','poutcome','y']:
    len=preprocessing.LabelEncoder().fit(data1[feature])
    data1[feature]=len.transform(data1[feature])
```

```python
data1
```

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(data1.loc[:,'age':'poutcome'], 
                                                    data1['y'], test_size=0.2, random_state=422)
```

```python
RFC1=RandomForestClassifier(n_estimators=100).fit(X_train1,y_train1)
print(RFC1.get_params)
print(RFC1.score(X_train1,y_train1))
RFC1.score(X_test1,y_test1)
```

```python
OptimalRFC1=RandomForestClassifier(class_weight='balanced_subsample',criterion='gini',max_depth=8,max_features=9,max_leaf_nodes=1920,
min_samples_split=3,n_estimators=50,oob_score=True,random_state=19070,min_samples_leaf=2, min_impurity_decrease=0.000000).fit(X_train1,y_train1)
print(OptimalRFC1.get_params)
print(OptimalRFC1.score(X_train1,y_train1))
RFC1.score(X_test1,y_test1)
```
