```python
#图的类型
#有向图
DG = nx.DiGraph() 
DG.add_weighted_edges_from([(1,2,0.5), (3,1,0.75), (1,4,0.3)]) # 添加带权值的边 
print(DG.degree(weight='weight'))#所有图中的节点与权重 
print(DG.out_degree(1))#节点1的输出个数
print (DG.out_degree(1, weight='weight'))#节点1的输出的权重为0.8，如果有amount的属性，那么这里可以改写weight='amount'.那么输出的将是金额总额。
print (list(DG.successors(1)) )# 返回结果为后续节点
print (list((DG.predecessors(1))))#返回结果为输入节点
```

```python
#Multigraphs允许重边
import matplotlib.pyplot as plt
fig,axes=plt.subplots(1,1,figsize=(6,4))
MG=nx.MultiGraph() 
MG.add_weighted_edges_from([(1,2,.95), (1,2,.95), (2,3,.0)]) #有重边
print(MG.degree(weight='weight'))#所有节点与权重 
GG=nx.Graph() 
print(list(MG.adjacency()))#返回一个二元元组列表，每个二元元组都由两部分组成，第一部分是节点名，第二部分是该节点周围的节点以及节点属性分布
for n,nbrs in MG.adjacency():     
    for nbr,edict in nbrs.items(): #在n节点的周围节点与属性间循环       
        minvalue=min([d['weight']for d in edict.values()])         
        GG.add_edge(n,nbr, weight = minvalue)  
print (nx.shortest_path(GG,1,3))
nx.draw(MG,ax=axes,with_labels=True)
#nx.draw(GG,ax=axes,with_lables=True,node_color = 'b',edge_color = 'r')
plt.show()
#dir(MG)
#注意：这里长度表示了权重,越近代表权重越高
```

```python
#物流线路图模拟
from scipy import linalg
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
fig,axes=plt.subplots(1,1,figsize=(14,12),dpi=180)
np.random.seed(3)
X=np.array(list(map(lambda x:'%.2f'% x,np.random.uniform(10,200,30))),dtype=np.float64).tolist()
np.random.seed(4)
Y=np.array(list(map(lambda x:'%.2f'% x,np.random.uniform(10,200,30))),dtype=np.float64).tolist()
t=list(set(list(zip(X,Y))))
n=len(t)
#print(n)
#G=nx.complete_graph(n)
#G=nx.Graph()#向量图
G=nx.DiGraph()#无向量图
#G[4][7]['weight']=123
label_list= range(0,n)
G.add_nodes_from(label_list)

label_dict={i:'S%u'%i for i in label_list}
pos={i:np.array(j) for i,j in zip(label_list,t)}
np.random.seed(5)
edgelist=list(zip(np.random.randint(0,n,100).tolist(),np.random.randint(0,n,100).tolist()))
edgelist=list(set(edgelist))
data=[]
for k1,k2 in edgelist:
    abst=linalg.norm(pos[k1]-pos[k2])
    data.append((k1,k2,int(abst)))
    
G.add_weighted_edges_from(data) #产生权重图的关键语句
#print(G.edges[1,19])
#list(G.adjacency())
nx.draw_networkx_nodes(G,pos,nodelist=None,node_size=300,node_color=None,
                        node_shape='o',alpha=1.0,cmap=None,vmin=None,
                        vmax=None,ax=None,linewidths=None,label=None)
nx.draw_networkx_edges(G,pos,width=1.0,edge_color='b',
                       style='solid',alpha=1.0,edge_cmap=None,edge_vmin=None,
                       edge_vmax=None,ax=None,arrows=True,label=None)

nx.draw_networkx_labels(G,pos,labels=label_dict)

edge_labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G, pos,edge_labels)
plt.show()
#添加边权重__________________________________________________________________

#pos=将节点作为键和位置作为值的字典。

print (nx.shortest_path(G,11,26,weight='weight'))#适合非权重图也适合权重图，权重图时不要忘记加weight关键字
print(nx.shortest_path_length(G,11,26,weight='weight'))#适合非权重图也适合权重图，权重图时不要忘记加weight关键字
print(nx.dijkstra_path_length(G,11,26))#只适合权重图
nx.dijkstra_path(G,11,26)#适合权重图
#list(nx.connected_components(G))#联通状况查询，适合无向量图
#nx.all_pairs_node_connectivity(G)#考察一个节点与其他节点的连接情况
```
